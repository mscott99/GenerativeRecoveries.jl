```julia
using Colors
using MLDatasets: MNIST
using FFTW: plan_r2r, REDFT00,  plan_dct, plan_idct, dct, fftshift, fft, ifft
using FileIO
using CairoMakie: heatmap
```

```julia
file = load("../datasets/celeba/000068.jpg")
```

```julia

img = convert(Array{Float32,2}, Gray.(file))
```

```julia
coeffs = fftshift(fft(img))
visible_coeffs = log.(abs.(coeffs) .+ 1f0)
visible_coeffs = visible_coeffs/maximum(visible_coeffs)
heatmap(visible_coeffs)
```
