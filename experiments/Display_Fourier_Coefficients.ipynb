```julia
using Colors
using MLDatasets: MNIST
using FFTW: plan_r2r, REDFT00,  plan_dct, plan_idct, dct
using FileIO
```

```julia
file = load("../datasets/celeba/000001.jpg")
img = convert(Array{Float32,2}, Gray.(file))
heatmap(img)
```

```julia
visible_coeffs = log.(abs.(dct(img)) .+ 1f0)
visible_coeffs = visible_coeffs/maximum(visible_coeffs)
heatmap(visible_coeffs)
```
